```python
import pandas as pd

# Step 1: Read the CSV file
input_file = 'Baseline_noise.csv'  # Đường dẫn file đầu vào
df = pd.read_csv(input_file)

# Step 2: Filter rows where noise_percentage is 0
filtered_df = df[df['noise_percentage'] == 0]

# Step 3: Sort by 'Dataset' then by 'scaled'
filtered_df = filtered_df.sort_values(by=['Dataset', 'scaled'])

# Step 4: Save the filtered and sorted rows to a new CSV file
output_file = 'Baseline_OC.csv'  # Đường dẫn file đầu ra
filtered_df.to_csv(output_file, index=False)

print(f"Dữ liệu đã lọc và sắp xếp được lưu vào '{output_file}'")
```
